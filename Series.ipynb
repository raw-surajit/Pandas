## Pandas Series

**A Series is a one-dimensional labled array capable of holding any data type. The axis lables are collectively called the index.**

## Creating Series

**There are multiple ways to create a pandas Series:**

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a', 'b', 'c']
my_list = [10,20,30]
arr = np.array([10,20,30])
d = {'a':10,'b':20,'c':30}
```

```python
pd.Series(my_list)
```

```python
pd.Series(my_list,index=labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(d)
```
